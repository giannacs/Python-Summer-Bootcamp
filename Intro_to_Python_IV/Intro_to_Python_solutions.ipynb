# Exercise 1
#### Queries
Perform the following queries
1. Select a subset of rows in the surveys_df DataFrame that contain data from the year 1999 and weights less than or equal to 8.
   How many rows did you end up with? What did your neighbor get?
2. Display **record_id** and **plot_id** of data that contain data of the years 1999, 2000, 1977 and 1989.
3. Write query that selects all rows with sex NOT equal to ‘M’ or ‘F’ in the data. Write two different queries that produce this same result. Write two different queries that do the same thing.

#### Solution to 1.1
A subset of rows in the surveys_df DataFrame that contain data from the year 1999 and weights less than or equal to 8.

```python
surveys_df[(surveys_df['year'] >= 1999) & (surveys_df['weight'] <= 8)]
```

How many rows did you end up with?<br>
Compare with a neighbour

```python
surveys_df[(surveys_df['year'] >= 1999) & (surveys_df['weight'] <= 8)].shape
```

#### Soltion to 1.2
**record_id** and **plot_id** of data that contain data of the years 1999, 2000, 1977 and 1989.

###### One solution

```python
surveys_df[surveys_df['year'].isin([1999, 2000, 1977, 1989])][['record_id','plot_id']]
```

###### Alternative solution

```python
surveys_df[['record_id','plot_id']][surveys_df['year'].isin([1999, 2000, 1977, 1989])]
```

#### Solution to 1.3
All rows with sex equal to ‘M’ or ‘F’ in the data. Write two different queries that produce this same result. Write two different queries that do the same thing.

###### One solution

```python
surveys_df[(surveys_df['sex'] == 'M') | (surveys_df['sex'] == 'F')]
```

###### Alternative solution

```python
surveys_df[surveys_df['sex'].isin(['M','F'])]
```

# Exercise 2
#### Plot
Create plots to answer th following questions
1. **Which location has the most animals?**<br>
   *Hints*:<br> 
   Create a plot of the number of animals per site (**plot_id**)<br>
   Group by **plot_id** and count the number of **record_id**s<br><br>
   
2. **Which species has the highest average weights?**<br>
   *Hints*:<br>
   Create a plot of average weight per species<br>
   Group by **species_id** and find the average (**mean**) of the **weight**s

```python
# Make sure figures appear inline in Ipython Notebook
%matplotlib inline
```

#### Solution to 2.1
Location wth most animals

```python
loc_counts = surveys_df.groupby('plot_id')['record_id'].count()
print(loc_counts)
loc_counts.plot(kind='bar')
```

###### Another example showing two bars per selection

```python
sex_means = surveys_df[['sex','hindfoot_length','weight']].groupby('sex')['hindfoot_length','weight'].mean()
print(sex_means)
sex_means.plot(kind='bar')
```

#### Solution to 2.2
Species with the highest average weights

```python
weight_avgs = surveys_df.groupby('species_id')['weight'].mean()
print(weight_avgs)
weight_avgs.plot(kind='bar')
```

Null values make the bar chart dirty

######  Use pie chart to make it clearer. The pie removes all null values

```python
weight_avgs.plot(kind='pie')
```

###### Remove nulls from dataframe first

```python
weight_avgs = surveys_df[~(surveys_df['weight'].isnull())].groupby('species_id')['weight'].mean()
print(weight_avgs)
weight_avgs.plot(kind='bar')
```

```python
weight_avgs = surveys_df[~(surveys_df['weight'].isnull())].groupby('species_id')['weight'].mean()
print(weight_avgs)
weight_avgs.plot(kind='pie')
```
